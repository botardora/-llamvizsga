```python
from geopy.geocoders import Nominatim
```

```python
geolocator = Nominatim(timeout=10, user_agent = "bi2023")
```

```python
import time
```

```python
from geopy.exc import GeocoderTimedOut
# Adott adattábla megyék oszlopa
megyek_oszlop = data[' Macroregiuni  regiuni de dezvoltare si judete']

# Megyék koordinátáinak tárolására szolgáló dictionary
koordinatak = {}

# Megpróbáljuk a geokódolást minden egyedi megye névével
for megye in megyek_oszlop.dropna().unique():
    # Megadjuk a keresett helyszínt, ami a megye neve Romániában
    helyszin = f'{megye}, Romania'

    try:
        # Geokódolás
        location = geolocator.geocode(helyszin)

        if location is not None:
            # Helyszín találata esetén hozzáadjuk a koordinátákat a dictionary-hoz
            koordinatak[megye] = (location.latitude, location.longitude)
            print(f'{megye} koordinátáinak lekérdezése megtörtént.')
        else:
            # Ha a helyszín nem található, hibaüzenetet nyomtatunk ki
            print(f'{megye} helyszín nem található.')
    except GeocoderTimedOut:
        # Időtúllépés esetén hibaüzenetet nyomtatunk ki
        print(f'{megye} helyszín geokódolása megszakadt időtúllépés miatt.')

# Eredmények kiírása
for megye, (lat, lon) in koordinatak.items():
    print(f'{megye}: Latitude={lat}, Longitude={lon}')
df_koordinatak = pd.DataFrame(koordinatak.items(), columns=['Megye', 'Koordinatak'])

# Adatok mentése CSV fájlba
df_koordinatak.to_csv('megye_koordinatak.csv', index=False)
print('Adatok mentése megtörtént.')
```

```python
df_merge = pd.merge(data, df_koordinatak, left_on=' Macroregiuni  regiuni de dezvoltare si judete', right_on='Megye', how='left')

# Az eredmény mentése
df_merge.to_csv('merged_table.csv')
```
