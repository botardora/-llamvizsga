```python
import pandas as pd
```

```python
df = pd.read_csv('alcoholan.csv')
```

```python
df
```

```python
columns_to=[' Unitati de masura', 'Unnamed: 0', 'Decedati']
```

```python
df=df.drop(columns=columns_to)
```

```python
df = df.rename(columns={'Principalele produse alimentare si bauturi': 'Termékek'})
df = df.rename(columns={' Ani': 'Év'})
df = df.rename(columns={' Valoare': 'Érték'})
```

```python
df['Termékek'] = df['Termékek'].str.replace('Vin si produse din vin', 'Bor és borból készült termékek')
df['Termékek'] = df['Termékek'].str.replace('Bere', 'Sör')
df['Termékek'] = df['Termékek'].str.replace('Bauturi alcoolice distilate (alcool 100%)', 'Desztillált alkoholok (100%-os alkohol)')
df['Termékek'] = df['Termékek'].str.replace('Consum total de alcool (alcool 100%)', 'Teljes alkoholfogyasztás (100%-os alkohol)')
```

```python
df = df.pivot(index='Év', columns='Termékek', values='Érték')
```

```python
df.to_csv('alkoholevmagyar.csv')
```
